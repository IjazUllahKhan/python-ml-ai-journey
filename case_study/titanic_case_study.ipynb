```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading titanic data set
titanic = sns.load_dataset("titanic")
titanic.head(5)
```

```python
# Saving Dataframe into csv file
titanic.to_csv("titanic.csv")
```

```python
# Basic statistics or summary
titanic.describe()
```

```python
titanic.head(5)
```

```python
# Dropping few columns and make a new Data Set
new_titanic = titanic.drop(["deck","embarked","alone"], axis = 1)
new_titanic.head()
```

```python
titanic.mean(numeric_only=1)
```

```python
titanic.groupby(["sex","class"]).mean(numeric_only= 1)
```

```python
titanic.value_counts("survived")
```

```python
titanic.groupby("sex").mean(numeric_only= 1)
```

```python
# Children in titanic
children = titanic[titanic["age"] < 18]
children.groupby("sex").mean(numeric_only=1)
```
