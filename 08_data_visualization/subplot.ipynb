# Subplots

```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("diamonds.csv")
df.head()
```

```python


fig,axes = plt.subplots(1,2,figsize=(12,5))
axes[0].hist(df["carat"],bins=10,edgecolor="black")
axes[0].set_title("Histogram of carat")
axes[1].hist(df["price"],bins=10,edgecolor="black")
axes[1].set_title("Histogram of price")
fig.suptitle("My fig title")
plt.show()
```
