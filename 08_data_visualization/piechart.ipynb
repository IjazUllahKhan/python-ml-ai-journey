```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating pie chart
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# Adding lables to pie chart
mylabels = np.array(["Apples","Bananas","Cherries","Dates"])
plt.pie(y,labels=mylabels)
plt.show()
```

```python
# Explode

myexplode = [0,0,0,0.3]
plt.pie(y,labels= mylabels , explode= myexplode)
plt.show()
```

```python
plt.pie(y,explode=myexplode,autopct="%1.1f%%")
plt.legend(title= "fruits Color:",labels= mylabels)
plt.show()
```
