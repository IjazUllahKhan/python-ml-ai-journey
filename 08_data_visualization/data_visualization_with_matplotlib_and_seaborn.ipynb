```python
# Loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
# line plot
plt.plot(df["carat"],df["price"])
plt.title("Carat vs price (line plot)")
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
# Scatter plot
plt.scatter(df["carat"],df["price"])
plt.title("cart vs price (scatter plot)")
plt.xlabel("carat")
plt.ylabel("price")
```
