```python
# Loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# line plot
plt.plot(df["carat"],df["price"])
plt.title("Carat vs price (line plot)")
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
# Scatter plot
plt.scatter(df["carat"],df["price"])
plt.title("cart vs price (scatter plot)")
plt.xlabel("carat")
plt.ylabel("price")
```

```python
# Bar plot
plt.bar(df["cut"].unique(),df.groupby("cut")["price"].mean())
plt.title("Average price by cut")
plt.xlabel("cut")
plt.ylabel("price")
```

```python
plt.hist(df["price"],bins=10, color="blue",edgecolor="black")
plt.title("Distribution of price")
plt.xlabel("price")
plt.ylabel("frequency")
```

```python
plt.boxplot(df["carat"])
plt.title("BOx plot of carat")
plt.ylabel("carat")
```

```python
cut_counts = df["cut"].value_counts()
plt.pie(cut_counts, labels= cut_counts.index,autopct='%1.1f%%')
plt.title("Distribution of cut Quality")
plt.show()
```
