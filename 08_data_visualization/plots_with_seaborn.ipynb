# Plot with seaborn

```python
# import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diamonds.csv")
df.head()
```

```python
sns.scatterplot(x = "carat", y = "price" , data = df , hue="cut" )
plt.title("Seabor scatter plot of carat vs price")
plt.show()
```

```python
# Seaborn line plot 
sns.lineplot(x = "carat",y="price", data=df)
plt.title("Carat vs price")
plt.show()
```

```python
# Box plot
sns.boxplot(y="price",x="cut",data=df)
plt.title("Box plot of cut")
plt.show()
```
